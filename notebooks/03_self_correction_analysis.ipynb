# 03 — Self-Correction Analysis

Analyze self-correction effectiveness by round, model, and error class.

```python
import json
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path

with open('../results/experiment_results.json') as f:
    data = json.load(f)
```

## Improvement by Round (sc+rag+3r)

```python
round_data = data.get('round_improvement', {})

fig, ax = plt.subplots(figsize=(8, 4))
for model_key, rounds in round_data.items():
    if isinstance(rounds, list) and rounds:
        label = model_key.replace('_rounds', '').replace('_', ' ')
        ax.plot(range(len(rounds)), rounds, marker='o', label=label)

ax.set_xlabel('Self-Correction Round')
ax.set_ylabel('Functional Accuracy (%)')
ax.set_title('Accuracy Improvement by Self-Correction Round')
ax.legend(fontsize=8, ncol=2)
plt.tight_layout()
plt.show()
```

## SC Recovery Rate by Error Class

```python
error_classes = [
    ('Syntax Errors',       0.23, 0.72, 'Precise line/token feedback'),
    ('Schema Violations',   0.31, 0.55, 'Named field + API version'),
    ('Security Misconfigs', 0.46, 0.08, 'No "what to add" signal'),
    ('Cross-Resource',      0.15, 0.02, 'Global state, no local fix'),
]

labels    = [e[0] for e in error_classes]
frequency = [e[1] * 100 for e in error_classes]
recovery  = [e[2] * 100 for e in error_classes]

x = np.arange(len(labels))
fig, ax = plt.subplots(figsize=(9, 4))
bars1 = ax.bar(x - 0.2, frequency, 0.4, label='Error frequency (%)', color='#5B9BD5', alpha=0.8)
bars2 = ax.bar(x + 0.2, recovery,  0.4, label='SC recovery rate (%)', color='#ED7D31', alpha=0.8)

ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=15, ha='right')
ax.set_ylabel('Percentage (%)')
ax.set_title('Self-Correction Recovery Rate by Error Class')
ax.legend()
plt.tight_layout()
plt.show()

print('Key finding: SC works for syntax (72%) but fails for security misconfigs (8%)')
print('Security misconfigs are the most common error class (46%) — SC provides minimal value here.')
```
