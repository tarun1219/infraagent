# 01 — Data Exploration

Explore the IaCBench benchmark structure and experiment results.

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

results_path = Path('../results/experiment_results.json')
with open(results_path) as f:
    data = json.load(f)

print(f'Top-level keys: {list(data.keys())}')
```

## Task Distribution

```python
# Task distribution across types and difficulties
dist = {
    'kubernetes': {1: 20, 2: 20, 3: 20, 4: 20, 5: 20},
    'terraform':  {1: 20, 2: 20, 3: 20, 4: 20, 5: 20},
    'dockerfile': {1: 20, 2: 20, 3: 20, 4: 20, 5: 20},
}
df = pd.DataFrame(dist).T
df.columns = [f'L{i}' for i in range(1, 6)]
df['Total'] = df.sum(axis=1)
print('IaCBench task distribution:')
print(df)
```

## Main Results Overview

```python
# Extract main results for DeepSeek
main = data.get('main_results', {})
rows = []
for key, val in main.items():
    if 'deepseek' in key:
        condition = key.replace('_deepseek', '')
        rows.append({
            'condition': condition,
            'functional': val.get('functional_accuracy', 0),
            'security': val.get('security_pass_rate', 0),
        })

df_main = pd.DataFrame(rows)
print('DeepSeek Coder v2 results:')
print(df_main.to_string(index=False))
```

```python
# Quick bar chart
fig, ax = plt.subplots(figsize=(8, 4))
x = range(len(df_main))
ax.bar([i - 0.2 for i in x], df_main['functional'], width=0.4, label='Functional')
ax.bar([i + 0.2 for i in x], df_main['security'],   width=0.4, label='Security')
ax.set_xticks(list(x))
ax.set_xticklabels(df_main['condition'], rotation=20, ha='right')
ax.set_ylabel('Pass Rate (%)')
ax.set_title('DeepSeek Coder v2 — Condition Comparison')
ax.legend()
plt.tight_layout()
plt.show()
```
